```python
import matplotlib.pyplot as plt
```

1.)Use matplotlib to plot the following equation:
y = x^2 - x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
plt.figure(figsize = (10,5))

x = range(-50,51)
y = [n ** 2 - n + 2 for n in x]
plt.title('$y=x^2 - x + 2$', fontsize = 14)
plt.ylabel('$y$', rotation = 0 )
plt.xlabel('$x$')
plt.plot(x,y)
plt.annotate('Origin', xy=(0, 0), xytext=(0, 250),arrowprops={'facecolor': 'blue'})


plt.show()
```

2.) Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

- y= sqrt(x)
- y = x^3
- y = 2^x
- y =1/(x +1)

```python
import math
```

```python
plt.figure(figsize = (10,5))

x = range(0,51)
y = [math.sqrt(n) for n in x]
plt.title('$y=\sqrt{x}$', fontsize = 14)
plt.xlabel('$x$')
plt.ylabel('$y=\sqrt{x}$')
plt.plot(x,y)
plt.show()



```

```python
plt.figure(figsize = (10,5))

x = range(0,51)
y = [n**3 for n in x]
plt.title('$y=x^3$', fontsize = 14)
plt.xlabel('$x$')
plt.ylabel("$x^3$", rotation = 0)
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize = (10,5))

x = range(0,51)
y = [2**n for n in x]
plt.title('$y=2^x$', fontsize = 14)
plt.ylabel('$2^x$')
plt.xlabel('$x$')
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize = (10,5))

x = range(0,51)
y = [1/(n+1) for n in x]
plt.title('$y = 1/(x+1)$', fontsize = 14)
plt.ylabel('$1/(x+1)$')
plt.xlabel('$x$')
plt.plot(x,y)
plt.show()
```

3.)Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x1 = range(0,51)
y = [math.sqrt(n) for n in x1]
y2 = [n**3 for n in x1]
y3 = [2**n for n in x1]
y4 = [1/(n+1) for n in x1]
```

```python
plt.figure(figsize = (10,10))

plt.suptitle('Y = X Fucntions', fontsize = 16)



plt.subplot(2,2,1) 
plt.plot(x1,y, ls = '-', marker = '*', color = 'red')
plt.title('$y=\sqrt{x}$')
plt.grid(True, ls = '-')

plt.subplot(2,2,2)
plt.plot(x1,y2, ls = '-', marker = 'v', color = 'purple')
plt.title('$y=x^3$')
plt.grid(True, ls = '-')

plt.subplot(2,2,3)
plt.plot(x1,y3, ls = '-', marker = 'o', color = 'blue')
plt.title('$y=2^x$')
plt.grid(True, ls = '-')

plt.subplot(2,2,4)
plt.plot(x1,y4, ls = '-', marker = '+', color = 'green')
plt.title('$1/(x+1)$')
plt.grid(True, ls = '-')


plt.tight_layout()
plt.show()
```

4.)Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x1 = [x * 0.25 for x in range(0,10)]
y = [math.sqrt(n) for n in x1]
y2 = [n**3 for n in x1]
y3 = [2**n for n in x1]
y4 = [1/(n+1) for n in x1]
```

```python
plt.figure(figsize = (10,10))

plt.suptitle('Y = X Fucntions', fontsize = 16)

plt.plot(x1,y, ls = '--', marker = '*', color = 'red',label='$y=\sqrt{x}$')
plt.plot(x1,y2, ls = '-', marker = 'v', color = 'purple', label='$y=x^3$')
plt.plot(x1,y3, ls = '--', marker = 'o', color = 'blue', label='$y=2^x$')
plt.plot(x1,y4, ls = '-', marker = '+', color = 'green', label='$y=1/(x+1)$')
plt.grid(True, ls = '-')
plt.legend(loc='upper left')








plt.show()
```
