```python
import matplotlib.pyplot as plt
import numpy as np 
import math
from random import randint
import random
```

```python
x = [x for x in range(1,4)]
y1 = [(0*n)+1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 =[n *(math.log(n))for n in x]
y5 =[n**2 for n in x]
y6 =[2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 =[n**n for n in x]

```

```python
plt.figure(figsize = (16,9))

plt.title('Big O Notation', fontsize = 16)# overall title for subplots
plt.xlabel('$Elements$', fontsize = 12)
plt.ylabel('$Operations$')
plt.plot(x,y1, ls = '-', marker = '*', color = 'red',label='$O(1)$')
plt.plot(x,y2, ls = '--', marker = 'o', color = 'blue',label='$O(log(n))$')
plt.plot(x,y3, ls = '-', marker = '^', color = 'orange',label='$O(n)$')
plt.plot(x,y4, ls = '--', marker = '|', color = 'purple',label='$O(n*log(n))$')
plt.plot(x,y5, ls = '-', marker = '*', color = 'green',label='$O(n^2)$')
plt.plot(x,y6, ls = '--', marker = 'v', color = 'yellow',label='$O(2^n)$')
plt.plot(x,y7, ls = '-', marker = '^', color = 'black',label='$O(n!)$')
plt.plot(x,y8, ls = '--', marker = 'o', color = 'green',label='$O(n^n)$')
plt.grid(True, ls = '-')

plt.legend(loc='upper left')



plt.show()
```
